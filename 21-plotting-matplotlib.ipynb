```python
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y
```

```python
plt.plot(X, Y)
```

```python
X.shape, Y.shape
```

```python
X, Y = np.meshgrid(X, Y)
X.shape, Y.shape
```

```python
plt.imshow(X, cmap="coolwarm")
```

```python
plt.imshow(Y, cmap="coolwarm")
```

```python
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)
```

```python
plt.imshow(R, cmap="coolwarm")
```

```python
plt.imshow(Z, cmap="coolwarm")
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='coolwarm')
```
